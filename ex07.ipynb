```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
yellow_sqr = cv2.imread('yellow_square.png', cv2.IMREAD_COLOR_RGB)
yellow_sqr.shape
```

```python
yellow_sqr = cv2.resize(yellow_sqr, (1000, 1000))
plt.imshow(yellow_sqr)
plt.title("Yellow Square")
plt.axis("on")
yellow_sqr.shape
```

```python
green_crcle = cv2.imread('green_circle.png', cv2.IMREAD_COLOR_RGB)
grey_crcle = cv2.cvtColor(green_crcle, cv2.COLOR_RGB2GRAY)
green_crcle.shape
```

```python
plt.figure(figsize = (8,6))
plt.imshow(green_crcle); plt.title("Green Circle")
```

```python
plt.imshow(grey_crcle, cmap = 'gray'); plt.title("Grayscale Image")
```

```python
ret, mask = cv2.threshold(grey_crcle, 127, 255, cv2.THRESH_BINARY)
```

```python
plt.figure(figsize = (16, 8))

plt.subplot(131); plt.imshow(green_crcle); plt.title("Green Circle")
plt.subplot(132); plt.imshow(grey_crcle, cmap = "gray"); plt.title("Grayscale Circle")
plt.subplot(133); plt.imshow(mask, cmap = "gray"); plt.title("Mask Image")
```

```python
mask_inv = cv2.bitwise_not(mask)
```

```python
plt.imshow(grey_crcle, cmap = "gray"); plt.title("Grayscale Circle")
```

```python
plt.imshow(mask_inv, cmap = "gray"); plt.title("Mask Inversed Image")
```

```python
plt.imshow(grey_crcle, cmap = "gray"); plt.title("Grayscale Circle")
```

```python
plt.imshow(mask, cmap = "gray"); plt.title("Mask Image")
```

```python
foreground = np.zeros_like(green_crcle) 
foreground = cv2.bitwise_and(foreground, foreground, mask=mask)
background = cv2.bitwise_and(yellow_sqr, yellow_sqr, mask = mask_inv)
result_image = cv2.add(foreground, background)
```

```python
plt.imshow(result_image)
plt.title("Resultant Image")
plt.axis("off")
```
